```python
STATUS_DATE_COL = "effective_date"
```

```python
status_history[STATUS_DATE_COL] = pd.to_datetime(status_history[STATUS_DATE_COL])
```

```python
status_history_sorted = status_history.sort_values(
    ["employee_id", STATUS_DATE_COL]
)

latest_status = (
    status_history_sorted
    .groupby("employee_id")
    .tail(1)
    .loc[:, ["employee_id", "status", STATUS_DATE_COL]]
    .rename(columns={
        "status": "current_status",
        STATUS_DATE_COL: "status_as_of"
    })
)

latest_status.head()
```

```python
termination_rows = status_history[
    status_history["status"].str.lower() == "terminated"
].copy()

termination_rows = termination_rows.sort_values(
    ["employee_id", STATUS_DATE_COL]
)

termination_dates = (
    termination_rows
    .groupby("employee_id")[STATUS_DATE_COL]
    .min()
    .rename("termination_date")
)

termination_dates.head()
```

```python
# Join employees with departments to get department_name
emp_dept = employees.merge(
    departments,
    on="department_id",
    how="left"
)

emp_dept.head()

```

```python
import pandas as pd

# --- 1) Clean snapshot date from JSON ---
snapshot_str = api_meta.get("extracted_at", None)

if snapshot_str is not None:
    # Parse as UTC, then drop timezone to make it timezone-naive
    snapshot_dt = pd.to_datetime(snapshot_str, utc=True).tz_convert(None)
else:
    snapshot_dt = pd.to_datetime("today").normalize()

# --- 2) Start from employees + departments ---
emp_dept = employees.merge(
    departments,
    on="department_id",
    how="left"
)

# --- 3) Merge in latest_status (Active / Terminated + status_as_of) ---
master_employee_df = emp_dept.merge(
    latest_status,
    on="employee_id",
    how="left"
)

# --- 4) Force hire_date and termination_date into clean datetime (no timezone) ---
master_employee_df["hire_date"] = pd.to_datetime(
    master_employee_df["hire_date"],
    utc=True
).dt.tz_convert(None)

master_employee_df["termination_date"] = pd.to_datetime(
    master_employee_df["termination_date"],
    utc=True,
    errors="coerce"     # this will keep NaT where there is no termination
).dt.tz_convert(None)

# --- 5) Build end_date: termination_date if present, else snapshot_dt ---
master_employee_df["end_date"] = master_employee_df["termination_date"].fillna(snapshot_dt)

# --- 6) Tenure metrics ---
master_employee_df["tenure_days"] = (
    master_employee_df["end_date"] - master_employee_df["hire_date"]
).dt.days

master_employee_df["tenure_years"] = master_employee_df["tenure_days"] / 365.25

master_employee_df.head()


```

```python
import numpy as np

# Make sure effective_date is clean and timezone-naive
status_history[STATUS_DATE_COL] = pd.to_datetime(
    status_history[STATUS_DATE_COL],
    utc=True
).dt.tz_convert(None)

# Add a month column (first day of the month)
status_history["month"] = status_history[STATUS_DATE_COL].dt.to_period("M").dt.to_timestamp()

# Classify each row as "hire", "termination", or "other"
status_history["event_type"] = np.where(
    status_history["status"].str.lower() == "terminated",
    "termination",
    np.where(
        status_history["status_reason"].str.lower() == "hire",
        "hire",
        "other"
    )
)

# Group by month and event_type
monthly_events = (
    status_history
    .groupby(["month", "event_type"])["employee_id"]
    .nunique()  # count distinct employees per event type per month
    .unstack(fill_value=0)
    .reset_index()
)

monthly_events.head()
```

```python
from pandas.tseries.offsets import MonthEnd

# Make sure our dates in master_employee_df are clean and timezone-naive
master_employee_df["hire_date"] = pd.to_datetime(
    master_employee_df["hire_date"],
    utc=True
).dt.tz_convert(None)

master_employee_df["termination_date"] = pd.to_datetime(
    master_employee_df["termination_date"],
    utc=True,
    errors="coerce"
).dt.tz_convert(None)

# Build a month range from first hire to snapshot
start_month = master_employee_df["hire_date"].min().to_period("M").to_timestamp()
end_month = snapshot_dt.to_period("M").to_timestamp()

month_index = pd.date_range(start=start_month, end=end_month, freq="MS")

headcount_rows = []

for m in month_index:
    month_start = m
    month_end = (m + MonthEnd(0))  # last day of the month
    
    # Employee is active AT END OF MONTH if:
    # hire_date <= month_end AND (no termination OR termination_date > month_end)
    active_mask = (
        (master_employee_df["hire_date"] <= month_end) &
        (
            master_employee_df["termination_date"].isna() |
            (master_employee_df["termination_date"] > month_end)
        )
    )
    
    headcount = active_mask.sum()
    
    headcount_rows.append({"month": month_start, "headcount_end": headcount})

headcount_by_month = pd.DataFrame(headcount_rows)

headcount_by_month.head()
```

```python
# Combine monthly events with headcount
monthly_churn = monthly_events.merge(
    headcount_by_month,
    on="month",
    how="left"
)

# If there is no headcount for a given month (shouldn't really happen), fill with 0
monthly_churn["headcount_end"] = monthly_churn["headcount_end"].fillna(0)

# For safety, ensure hire/termination columns exist
if "hire" not in monthly_churn.columns:
    monthly_churn["hire"] = 0
if "termination" not in monthly_churn.columns:
    monthly_churn["termination"] = 0

# Churn rate = terminations during month / headcount at end of month
monthly_churn["churn_rate"] = monthly_churn.apply(
    lambda row: row["termination"] / row["headcount_end"]
    if row["headcount_end"] > 0 else 0,
    axis=1
)

# Sort by month
monthly_churn = monthly_churn.sort_values("month")

monthly_churn.tail(12)
```
